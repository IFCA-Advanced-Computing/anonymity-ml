## Plot accuracy and AUC obtained when varying the value of $k$.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
knn_acc = [0.8056, 0.8176, 0.8199, 0.8099, 0.8153, 0.8095, 0.8170, 0.8124, 0.8129, 0.8107]
rf_acc = [0.8322, 0.8196, 0.8189, 0.8125, 0.8145, 0.8116, 0.8163, 0.8114, 0.8129, 0.8103]
gb_acc = [0.8352, 0.8249, 0.8221, 0.8168, 0.8149, 0.8097, 0.8177, 0.8114, 0.8129, 0.8103]
ab_acc = [0.8350, 0.8266, 0.8197, 0.8146, 0.8129, 0.8090, 0.8164, 0.8051, 0.8091, 0.8047]
```

```python
knn_auc = [0.6779, 0.7143, 0.7234, 0.6932, 0.7151, 0.6947, 0.7194, 0.7268, 0.7064, 0.7254]
rf_auc = [0.7342, 0.7167, 0.7123, 0.7063, 0.7067, 0.7173, 0.6987, 0.7285, 0.7054, 0.7274]
gb_auc = [0.7437, 0.7346, 0.7258, 0.7157, 0.7085, 0.7098, 0.7139, 0.7285, 0.7030, 0.7274]
ab_auc = [0.7377, 0.7372, 0.7342, 0.7171, 0.7152, 0.7120, 0.7167, 0.7399, 0.7195, 0.7117]
```

```python
labels = ['Raw', 'k=2', 'k=5', 'k=10', 'k=15', 'k=20', 'k=25', 'k=50', 'k=75', 'k=100']
```

```python
fig, axs = plt.subplots(2, 1, figsize = (8,8.5))
axs[0].plot(labels, knn_acc, label = 'kNN', color = '#2a40cb',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot(labels, rf_acc, label = 'RF', color = '#e23d15',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot(labels, gb_acc, label = 'GB', color = '#11bf2e',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot(labels, ab_acc, label = 'AB', color = '#c2cd0f',
        marker = "D", markerfacecolor = "white", markersize = 5)

axs[0].set_ylabel('Accuracy')
axs[0].set_title('Accuracy obtained with each ML model when varying the value of $k$')
axs[0].legend(loc = 0)

axs[1].plot(labels, knn_auc, label = 'kNN', color = '#2a40cb',
        marker = "s", markersize = 5)
axs[1].plot(labels, rf_auc, label = 'RF', color = '#e23d15',
        marker = "s", markersize = 5)
axs[1].plot(labels, gb_auc, label = 'GB', color = '#11bf2e',
        marker = "s", markersize = 5)
axs[1].plot(labels, ab_auc, label = 'AB', color = '#c2cd0f',
        marker = "s", markersize = 5)

axs[1].set_ylabel('AUC')
axs[1].set_title('AUC obtained with each ML model when varying the value of $k$')
axs[1].legend(loc = 0)

plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize = (10,10))
x = [-1, 2, 5, 10, 15, 20, 25, 50, 75, 100]
axs[0].plot(x, knn_acc, label = 'kNN', color = '#2a40cb',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot(x, rf_acc, label = 'RF', color = '#e23d15',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot(x, gb_acc, label = 'GB', color = '#11bf2e',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot(x, ab_acc, label = 'AB', color = '#c2cd0f',
        marker = "D", markerfacecolor = "white", markersize = 5)
axs[0].plot([x[0]], [rf_acc[0]], 'o', color = 'k')
axs[0].plot([x[0]], [knn_acc[0]], 'o', color = 'k')
axs[0].plot([x[0]], [gb_acc[0]], 'o', color = 'k')
axs[0].plot([x[0]], [ab_acc[0]], 'o', color = 'k')

axs[0].set_ylabel('Accuracy')
axs[0].set_xticks(x)
axs[0].set_xticklabels(labels, rotation=45)
axs[0].set_title('Accuracy obtained with each ML model when varying the value of $k$')
axs[0].legend(loc = 0)

axs[1].plot(x, knn_auc, label = 'kNN', color = '#2a40cb',
        marker = "s", markersize = 5)
axs[1].plot(x, rf_auc, label = 'RF', color = '#e23d15',
        marker = "s", markersize = 5)
axs[1].plot(x, gb_auc, label = 'GB', color = '#11bf2e',
        marker = "s", markersize = 5)
axs[1].plot(x, ab_auc, label = 'AB', color = '#c2cd0f',
        marker = "s", markersize = 5)

axs[1].set_ylabel('AUC')
axs[1].set_xticks(x)
axs[1].set_xticklabels(labels, rotation=45)
axs[1].set_title('AUC obtained with each ML model when varying the value of $k$')
axs[1].legend(loc = 0)
axs[1].plot([x[0]], [rf_auc[0]], 'o', color = 'k')
axs[1].plot([x[0]], [knn_auc[0]], 'o', color = 'k')
axs[1].plot([x[0]], [gb_auc[0]], 'o', color = 'k')
axs[1].plot([x[0]], [ab_auc[0]], 'o', color = 'k')

plt.show()
```
