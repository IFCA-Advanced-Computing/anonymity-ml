## Plot ROC curves and show AUC.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import auc
```

```python
roc_raw = pd.read_csv('roc_raw.csv')
roc_k5 = pd.read_csv('roc_k5.csv')
roc_k5_l2 = pd.read_csv('roc_k5_l2.csv')
roc_k5_t07 = pd.read_csv('roc_k5_t07.csv')
roc_k5_beta15 = pd.read_csv('roc_k5_beta15.csv')
```

```python
fpr_raw_knn = [0, roc_raw.kNN[0], 1.0]
tpr_raw_knn = [0, roc_raw.kNN[1], 1.0]

fpr_k5_knn = [0, roc_k5.kNN[0], 1.0]
tpr_k5_knn = [0, roc_k5.kNN[1], 1.0]

fpr_k5_l2_knn = [0, roc_k5_l2.kNN[0], 1.0]
tpr_k5_l2_knn = [0, roc_k5_l2.kNN[1], 1.0]

fpr_k5_t07_knn = [0, roc_k5_t07.kNN[0], 1.0]
tpr_k5_t07_knn = [0, roc_k5_t07.kNN[1], 1.0]

fpr_k5_beta15_knn = [0, roc_k5_beta15.kNN[0], 1.0]
tpr_k5_beta15_knn = [0, roc_k5_beta15.kNN[1], 1.0]
```

```python
fpr_raw_rf = [0, roc_raw.RF[0], 1.0]
tpr_raw_rf = [0, roc_raw.RF[1], 1.0]

fpr_k5_rf = [0, roc_k5.RF[0], 1.0]
tpr_k5_rf = [0, roc_k5.RF[1], 1.0]

fpr_k5_l2_rf = [0, roc_k5_l2.RF[0], 1.0]
tpr_k5_l2_rf = [0, roc_k5_l2.RF[1], 1.0]

fpr_k5_t07_rf = [0, roc_k5_t07.RF[0], 1.0]
tpr_k5_t07_rf = [0, roc_k5_t07.RF[1], 1.0]

fpr_k5_beta15_rf = [0, roc_k5_beta15.RF[0], 1.0]
tpr_k5_beta15_rf = [0, roc_k5_beta15.RF[1], 1.0]
```

```python
fpr_raw_gb = [0, roc_raw.GB[0], 1.0]
tpr_raw_gb = [0, roc_raw.GB[1], 1.0]

fpr_k5_gb = [0, roc_k5.GB[0], 1.0]
tpr_k5_gb = [0, roc_k5.GB[1], 1.0]

fpr_k5_l2_gb = [0, roc_k5_l2.GB[0], 1.0]
tpr_k5_l2_gb = [0, roc_k5_l2.GB[1], 1.0]

fpr_k5_t07_gb = [0, roc_k5_t07.GB[0], 1.0]
tpr_k5_t07_gb = [0, roc_k5_t07.GB[1], 1.0]

fpr_k5_beta15_gb = [0, roc_k5_beta15.GB[0], 1.0]
tpr_k5_beta15_gb = [0, roc_k5_beta15.GB[1], 1.0]
```

```python
fpr_raw_ab = [0, roc_raw.AB[0], 1.0]
tpr_raw_ab = [0, roc_raw.AB[1], 1.0]

fpr_k5_ab = [0, roc_k5.AB[0], 1.0]
tpr_k5_ab = [0, roc_k5.AB[1], 1.0]

fpr_k5_l2_ab = [0, roc_k5_l2.AB[0], 1.0]
tpr_k5_l2_ab = [0, roc_k5_l2.AB[1], 1.0]

fpr_k5_t07_ab = [0, roc_k5_t07.AB[0], 1.0]
tpr_k5_t07_ab = [0, roc_k5_t07.AB[1], 1.0]

fpr_k5_beta15_ab = [0, roc_k5_beta15.AB[0], 1.0]
tpr_k5_beta15_ab = [0, roc_k5_beta15.AB[1], 1.0]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize = (7, 6))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_raw_knn, tpr_raw_knn, 
         label='Raw data (AUC = {:.4f})'.format(auc(fpr_raw_knn, tpr_raw_knn)))
plt.plot(fpr_k5_knn, tpr_k5_knn, 
         label='k=5 (AUC = {:.4f})'.format(auc(fpr_k5_knn, tpr_k5_knn)))
plt.plot(fpr_k5_l2_knn, tpr_k5_l2_knn, 
         label='k=5, $\ell=2$ (AUC = {:.4f})'.format(auc(fpr_k5_l2_knn, tpr_k5_l2_knn)))
plt.plot(fpr_k5_t07_knn, tpr_k5_t07_knn, 
         label='k=5, $t=0.7$ (AUC = {:.4f})'.format(auc(fpr_k5_t07_knn, tpr_k5_t07_knn)))
plt.plot(fpr_k5_beta15_knn, tpr_k5_beta15_knn, 
         label='k=5, $=1.5$ (AUC = {:.4f})'.format(auc(fpr_k5_beta15_knn, tpr_k5_beta15_knn)))
plt.xlabel('FPR (false positive rate)')
plt.ylabel('TPR (true positive rate)')
plt.title('ROC curves. kNN case.')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize = (7, 6))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_raw_rf, tpr_raw_rf, 
         label='Raw data (AUC = {:.4f})'.format(auc(fpr_raw_rf, tpr_raw_rf)))
plt.plot(fpr_k5_rf, tpr_k5_rf, 
         label='k=5 (AUC = {:.4f})'.format(auc(fpr_k5_rf, tpr_k5_rf)))
plt.plot(fpr_k5_l2_rf, tpr_k5_l2_rf, 
         label='k=5, $\ell=2$ (AUC = {:.4f})'.format(auc(fpr_k5_l2_rf, tpr_k5_l2_rf)))
plt.plot(fpr_k5_t07_rf, tpr_k5_t07_rf, 
         label='k=5, $t=0.7$ (AUC = {:.4f})'.format(auc(fpr_k5_t07_rf, tpr_k5_t07_rf)))
plt.plot(fpr_k5_beta15_rf, tpr_k5_beta15_rf, 
         label='k=5, $=1.5$ (AUC = {:.4f})'.format(auc(fpr_k5_beta15_rf, tpr_k5_beta15_rf)))
plt.xlabel('FPR (false positive rate)')
plt.ylabel('TPR (true positive rate)')
plt.title('ROC curves. RF case.')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize = (7, 6))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_raw_gb, tpr_raw_gb, 
         label='Raw data (AUC = {:.4f})'.format(auc(fpr_raw_gb, tpr_raw_gb)))
plt.plot(fpr_k5_gb, tpr_k5_gb, 
         label='k=5 (AUC = {:.4f})'.format(auc(fpr_k5_gb, tpr_k5_gb)))
plt.plot(fpr_k5_l2_gb, tpr_k5_l2_gb, 
         label='k=5, $\ell=2$ (AUC = {:.4f})'.format(auc(fpr_k5_l2_gb, tpr_k5_l2_gb)))
plt.plot(fpr_k5_t07_gb, tpr_k5_t07_gb, 
         label='k=5, $t=0.7$ (AUC = {:.4f})'.format(auc(fpr_k5_t07_gb, tpr_k5_t07_gb)))
plt.plot(fpr_k5_beta15_gb, tpr_k5_beta15_gb, 
         label='k=5, $=1.5$ (AUC = {:.4f})'.format(auc(fpr_k5_beta15_gb, tpr_k5_beta15_gb)))
plt.xlabel('FPR (false positive rate)')
plt.ylabel('TPR (true positive rate)')
plt.title('ROC curves. GB case.')
plt.legend(loc='best')
plt.show()
```
